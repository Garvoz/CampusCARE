```python
import numpy as np
import pandas as pd
import pickle
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("../donnees/nettoyees/student_depression_dataset_nettoyees.csv")
```

```python
df
```

```python
#Déclaration des variables X et y
X = df.drop(columns = ['Dépression', 'Pensées Suicidaires', 'Antécédents familiaux mentaux'])#On drop les colonnes de booléens pour ne pas les standardiser. On les concaténera plus tard
y = df['Dépression']
X_reserve = df[['Pensées Suicidaires', 'Antécédents familiaux mentaux']]
```

Ici on va chercher à faire une Classification car un veut savoir si la colonne dépression sera de 0 ou 1.
Il faut donc faire une standardisation des données

```python
Xnum = X[['Age', 'Pression Académique', 'CGPA', 'Satisfation Académique', 'Temps études/jour(heures)', 'Stress Financier']]
Xcat = X[['Genre', 'Ville', 'Niveau étude', 'Temps de sommeil', 'Habitudes Alimentaires']]
```

```python
#Standardisation des données numériques
SN = StandardScaler()
Xnum_SN = pd.DataFrame(SN.fit_transform(Xnum), columns=Xnum.columns)
Xnum_SN['CGPA'] = Xnum_SN['CGPA'] * 2#Donnons un poids *2 à la colonne CGPA
Xnum_SN
```

```python
#get_dummies sur les données catégorielles
Xcat_dummies = pd.get_dummies(Xcat)
Xcat_dummies
```

```python
#Concaténation finale des colonnes
X_final = pd.concat([Xnum_SN, X_reserve], axis=1)
X_final = pd.concat([X_final, Xcat_dummies], axis=1)
```

```python
X_final
```

```python
#Mise en place des variables pour le train test split:
X_train, X_test, y_train, y_test = train_test_split(X_final, y, train_size=0.8, random_state=12)
```

```python
#Déclaration et entrainement du modèle sur X_train et y_train:
model = LogisticRegression()

model.fit(X_train, y_train)
```

```python
#Affichage des scores d'entrainement et de test:
print(f"Score d'entrainement {model.score(X_train, y_train)}")
print(f"Score de test {model.score(X_test, y_test)}")
#Le modèle est assez performant et très robuste.
```

```python
y_pred = model.predict(X_test)
pd.crosstab(y_test, y_pred, rownames=['Dépression réelle'], colnames=['Dépression prédite'])
```

Environ 85% de bonnes prédictions. Modèle assez performant mais pas assez pour une prédiction de dépression. On favorisera donc un résultat en pourcentage de probabilité


On va maintenant faire une fonction qui va prédire si il y dépression pour une nouvelle entrée.

```python
#On utilisera predict_proba plutôt que predict pour avoir un pourcentage de probabilité de dépression plutôt qu'une certitude
probabilities = model.predict_proba(X_test)

# Affichons par exemple les probabilités pour les 5 premieres lignes de X_test:(array avec d'abord proba de 0 puis de 1)
print(probabilities[:5])
```

```python
#Définition des paramêtres pour le ML:
g = 'Homme'
a = 34
v = 'Vadodara'
niv = 'Bachelor'
pa = 1
cgpa = 6.21
sa = 2
ts = '5-6 heures'
ha = 'Mauvaises'
ps = False
tej = 8
sf = 5
afm = True
```

```python
#On met ces parametres dans un dico pour en faire une ligne de df similaire au df de départ eet ainsi pouvoir prédire la dépression:
dico_a_remplir = {
    'Genre' : g,
    'Age' : a, 
    'Ville' : v, 
    'Niveau étude' : niv, 
    'Pression Académique' : pa, 
    'CGPA' : cgpa,
    'Satisfation Académique' : sa,
    'Temps de sommeil' : ts, 
    'Habitudes Alimentaires' : ha,
    'Pensées Suicidaires' : ps, 
    'Temps études/jour(heures)' : tej, 
    'Stress Financier' : sf,
    'Antécédents familiaux mentaux' : afm
}
```

```python
df_predict = pd.DataFrame([dico_a_remplir])
```

```python
df_predict
```

```python
#Fonction pour encoder le df_predict comme le X_final:
def encodage_predict(df_a_predire):
    X_num = df_a_predire[['Age', 'Pression Académique', 'CGPA', 'Satisfation Académique', 'Temps études/jour(heures)', 'Stress Financier']]
    X_cat = df_a_predire[['Genre', 'Ville', 'Niveau étude', 'Temps de sommeil', 'Habitudes Alimentaires']]
    X_reserve = df_a_predire[['Pensées Suicidaires', 'Antécédents familiaux mentaux']]

    X_num_SN = pd.DataFrame(SN.transform(X_num), columns=X_num.columns)

    X_cat_dummies = pd.get_dummies(X_cat)
    X_encoded_predire = pd.concat([X_num_SN, X_reserve], axis=1)
    X_encoded_predire = pd.concat([X_encoded_predire, X_cat_dummies], axis=1)

    # DataFrame vide qui a les mêmes colonnes que X_final
    X_predict = pd.DataFrame(columns=X_final.columns)

    # On veut que le DataFrame ait le même nombre de lignes que X_encoded_predire
    X_predict = X_predict.reindex(index=X_encoded_predire.index)
    # On met tous les NaN à False
    X_predict = X_predict.fillna(False)

    # On parcourt chaque colonne de X_encoded_predire
    # Si la colonne est présente dans X_final alors on la garde
    # Sinon, on la met à False
    for column in X_final.columns:
        if column in X_encoded_predire.columns:
            X_predict[column] = X_encoded_predire[column]

    return X_predict

    
```

```python
predict = encodage_predict(df_predict)
```

```python
predict
```

```python
#Fonction qui sort le pourcentage de risque que la personne soit effectivement dépressive:
proba = float(np.round(model.predict_proba(predict)[0, 1] * 100, 2))
proba
```
